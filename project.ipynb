### Group Dishes Based on Ingredients or States

```python
from sklearn.feature_extraction.text import CountVectorizer

# Convert ingredients to vectors using CountVectorizer
vectorizer = CountVectorizer(stop_words='english')
ingredient_matrix = vectorizer.fit_transform(df['Ingredients'].fillna(''))

# Display the shape of the matrix (number of dishes x number of unique ingredients)
print(ingredient_matrix.shape)
```

### Apply K-Means Clustering:



```python
from sklearn.cluster import KMeans

# Apply KMeans clustering (let’s say we want to group into 5 clusters)
kmeans = KMeans(n_clusters=5, random_state=42)
df['Cluster'] = kmeans.fit_predict(ingredient_matrix)

# Show the first few rows of the DataFrame with clusters
print(df[['Dish', 'State', 'Cluster']].head())
```

### Visualizing Clusters:

```python
from sklearn.decomposition import PCA

# Reduce dimensions using PCA
pca = PCA(n_components=2)
pca_components = pca.fit_transform(ingredient_matrix.toarray())

# Plot the clusters
plt.figure(figsize=(8, 6))
plt.scatter(pca_components[:, 0], pca_components[:, 1], c=df['Cluster'], cmap='viridis', marker='o')
plt.title("Dishes Clustering (PCA projection)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.colorbar()
plt.show()
```

### Classification Models (Supervised Learning)

```python
# Assuming there is a column "Vegetarian" in your dataset (1 = vegetarian, 0 = non-vegetarian)
# If it's not present, you can create a target variable based on ingredients.

df['Vegetarian'] = df['Ingredients'].apply(lambda x: 1 if 'vegetarian' in x.lower() else 0)

# Split into features and target
X = df['Ingredients']  # Features (ingredients)
y = df['Vegetarian']    # Target (vegetarian or not)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Convert ingredients to numerical features
vectorizer = TfidfVectorizer(stop_words='english')
X_vec = vectorizer.fit_transform(X)
```

### Train a Classification Model:



```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_vec, y, test_size=0.2, random_state=42)

# Train a Random Forest classifier
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)

# Make predictions
y_pred = clf.predict(X_test)

# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:")
print(classification_report(y_test, y_pred))
```
